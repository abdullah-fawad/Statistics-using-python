```python
Q1 = df['TotalWorkingYears'].quantile(0.25)
Q2 = df['TotalWorkingYears'].quantile(0.5)
Q3 = df['TotalWorkingYears'].quantile(0.75)

IQR = Q3 - Q1

Upper_threshhold = Q1 - 1.5 * IQR 
lower_threshold = Q3 + 1.5 * IQR

outlier=(df['TotalWorkingYears'] < Upper_threshhold) | (df['TotalWorkingYears'] > lower_threshold)
outlier
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(data=df, y='TotalWorkingYears')
plt.title('Box plot with Quartiles,IQR,Range,and Outliers')
plt.show()
```

```python
# The first line this is the minimum line, then the bottom of the box is the the Q1 the minimum range and the line in the middle of the box is
# Q2 also known as the median the top of the box is Q3 and the whole box is the IQR. The most top line is the maximum line. 
# And the points outside the whiskers are outliers as can be shown in the graph.  
```

```python
from scipy.stats import zscore

df['TotalWorkingYears_zscore'] = zscore(df['TotalWorkingYears'])
```

```python
# Z-score = 0: The data point is exactly equal to the mean.
# A z-score of +1 indicates the value is 1 standard deviation above the mean, and a z-score of -1 means it's 1 standard deviation below the mean.
# f a z-score is below -3 or above +3, it might be considered an outlier.
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df['MonthlyIncome'].corr(df['TotalWorkingYears'])
```

```python
# there is a strong positive linear relationship between monthly income and total working years which means that as people's experience 
# increases there income also increase.
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['TotalWorkingYears'], df['MonthlyIncome'], color='blue')
plt.title('Monthly Income vs Total Working Years')
plt.xlabel('Total Working Years')
plt.ylabel('Monthly Income')
plt.show()
```

```python
X = df['TotalWorkingYears'].values.reshape(-1, 1)
Y = df['MonthlyIncome']
model = LinearRegression()
model.fit(X,Y)

intercept = model.intercept_
slope = model.coef_[0]

print(f"Regression Line: y = {intercept:.2f} + {slope:.2f}x")
```

```python
#  1238.30 is the intercept this is the income of the employee having zero experience.
# 466.83 is the slope which is positive means that as the work experience increases income increases by 466.83.
```

```python
df['PredictedMonthlyIncomebyworkingyears'] = model.predict(X)
```

```python
y_pred = model.predict(X)
plt.figure(figsize = (7,4))
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.scatter(X, y_pred, color='red', label='Predicted Regression Line')
plt.title('Predicted Monthly Income vs Total Working Years')
plt.xlabel('Total Working Years')
plt.ylabel('Monthly Income')
plt.legend()
plt.show()
```
