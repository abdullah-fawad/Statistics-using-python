```python
# the graph shows us the employees job satisfaction with their monthly income. And we can see that the all bars are approaximately same with bar 1 being sligtly higher than the other bars.
```

```python
Q1 = df['JobSatisfaction'].quantile(0.25)
Q2 = df['JobSatisfaction'].quantile(0.5)
Q3 = df['JobSatisfaction'].quantile(0.75)

IQR = Q3 - Q1

Upper_threshhold = Q1 - 1.5 * IQR 
lower_threshold = Q3 + 1.5 * IQR

outlier=(df['JobSatisfaction'] < Upper_threshhold) | (df['JobSatisfaction'] > lower_threshold)
outlier
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(data=df, y='JobSatisfaction')
plt.title('Box plot with Quartiles,IQR,Range,and Outliers')
plt.show()
```

```python
# No oultiers in the Job Satisfaction column as their are no points outside the whiskers as shown in the graph.The center line of the box showing
# the median or Q2. 
```

```python
df['Job_satisfaction_zscore'] = zscore(df['JobSatisfaction'])
```

```python
df.head()
```

```python
df.to_csv('Update_2.csv', index=False)
```

```python
# the negative z score indicates that the it is lower then average monthly income and a positive z score indiactes that the standard deviations is above the averge monthly income
```

```python
df['MonthlyIncome'].corr(df['JobSatisfaction'])
```

```python
# this indicates a very week corelation between job satisfaction and monthly income as their are other factors also involved that affects job satisfaction>
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['JobSatisfaction'], df['MonthlyIncome'], color='navy')
plt.title('Monthly Income vs JobSatisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Monthly Income')
plt.show()
```

```python
X = df['JobSatisfaction'].values.reshape(-1, 1)
Y = df['MonthlyIncome']
model = LinearRegression()
model.fit(X,Y)

intercept = model.intercept_
slope = model.coef_[0]

print(f"Regression Line: y = {intercept:.2f} + ({slope:.2f}x)")
```

```python
df['PredictedMonthlyIncomebyJobsatisfaction'] = model.predict(X)
```

```python
df.to_csv('Update_2.csv', index=False)
```

```python
y_pred = model.predict(X)
plt.figure(figsize = (7,4))
plt.scatter(X, Y, color='red', label='Actual Data')
plt.scatter(X, y_pred, color='black', label='Predicted Regression Line')
plt.title('Predicted Monthly Income vs Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Monthly Income')
plt.legend()
plt.show()
```

```python
predictions = model.predict(X)
r2 = r2_score(Y,predictions)
```

```python
r2
```

```python
plt.scatter(df['JobSatisfaction'], df['MonthlyIncome'],c ='navy')
plt.xticks([1, 2, 3, 4, 5])

slope,intercept,rvalue,pvalue,stderr = stats.linregress(df['JobSatisfaction'], df['MonthlyIncome'])

plt.plot(df['JobSatisfaction'],slope * df['JobSatisfaction'] + intercept,color = 'crimson')
plt.annotate("y=%.3fx+%.3f\n"%(slope,intercept),xy=(0.2,0.4),xycoords=('figure fraction'),color ='black')
plt.title('Predicted Monthly Income vs Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Monthly Income')
plt.show()
```
