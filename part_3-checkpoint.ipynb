```python
from sklearn.linear_model import LinearRegression
```

```python
df['MonthlyIncome'].corr(df['Education'])
```

```python
# Monthly income and education level has a week linear relationship between each other. they both are connected with eachother but still their connection isn't that strong.
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Education'], df['MonthlyIncome'], color='skyblue')
plt.title('Monthly Income vs Education level')
plt.xlabel('Education level')
plt.ylabel('Monthly Income')
plt.show()
```

```python
X = df['Education'].values.reshape(-1, 1)
Y = df['MonthlyIncome']
model = LinearRegression()
model.fit(X,Y)

intercept = model.intercept_
slope = model.coef_[0]

print(f"Regression Line: y = {intercept:.2f} + {slope:.2f}x")
```

```python
# 5273.10 is the income of the employee having zero education level. and as the education level increases the income increases by 423.21.
```

```python
plt.scatter(df['Education'], df['MonthlyIncome'],c ='skyblue')
plt.xticks([1, 2, 3, 4, 5])

slope,intercept,rvalue,pvalue,stderr = stats.linregress(df['Education'], df['MonthlyIncome'])

plt.plot(df['Education'],slope * df['Education'] + intercept,color = 'crimson')
plt.annotate("y=%.3fx+%.3f\n"%(slope,intercept),xy=(0.5,0.2),xycoords=('figure fraction'),color ='black')
plt.title('Predicted Monthly Income vs Education level')
plt.xlabel('Education level')
plt.ylabel('Monthly Income')
plt.show()
```

```python
df['PredictedMonthlyIncomebyeduactionlevel'] = model.predict(X)
```

```python
df.head()
```

```python
y_pred = model.predict(X)
plt.figure(figsize = (7,4))
plt.scatter(X, Y, color='skyblue', label='Actual Data')
plt.scatter(X, y_pred, color='red', label='Predicted Regression Line')
plt.title('Predicted Monthly Income vs Education level')
plt.xlabel('Education level')
plt.ylabel('Monthly Income')
plt.legend()
plt.show()
```

```python
df.to_csv('Update_2.csv', index=False)
```

```python
from sklearn.metrics import r2_score
```

```python
predictions = model.predict(X)
r2 = r2_score(Y,predictions)
```
